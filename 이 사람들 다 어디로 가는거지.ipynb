### Q) 요일을 추가 할 수있을까?

```python
# 일시를 관리하는 datetie 라이브러리를 불러오기

from datetime import datetime
```



datetime.strptime('날짜str', `str형태`)  : 문자 --> 날짜 타입

- `str형태` %Y-%m-%d %H:%M:%S
    - %Y : 연도(4자리)
    - %m:  월(2자리)
    - %d:  일자(2자리)
    
    - %H : 시간
    - %M : 분
    - %S : 초

```python
# datetime.strptime 을 이용해 str타입을 날짜 타입으로 변경
# datetime.strptime('날짜str', 형태)

s = '20190630'     #int 가 아니라 str형태여야만함
date = datetime.strptime(s, '%Y%m%d')
date
```

```python
# 날짜 타입을 요일로 변경하기 위해서는 weekday 쓰기
# 월요일 :0 ~ 일요일 :6

weekday_dict = ['월', '화', '수', '목', '금', '토', '일']
weekday_index = date.weekday()
print(weekday_index)
print(weekday_dict[weekday_index])
```

```python
# 닐짜 컬럼을 불러와서, 순서대로 요일을 점검하여 리스트에 저장하기

weekday_dict = ['월', '화', '수', '목', '금', '토', '일']
weekday_list = []

for date_str in raw['사용일자']:
    
    date = datetime.strptime(str(date_str), "%Y%m%d")
    weekday_index = date.weekday()
    weekday = weekday_dict[weekday_index]
    
    weekday_list.append(weekday)
```

```python
# 요일정보가 저장된 리스트를 컬럼에 추가

raw['요일'] = weekday_list

raw.head()
```

## Q 5월에 승객수가 많은가?  
월별 승객수 비교 -> 필요한 컬럼추가(연월)

```python
raw.head()
```

```python
# 연월일 데이터에서 연월/월일 데이터 뽑기(중요)

yearmonth_list=[]
monthday_list=[]

for date in raw['사용일자']:
    yearmonth = str(date)[:6] # int' object is not subscriptable -> str로 변환
    yearmonth_list.append(yearmonth)
    monthday = str(date)[4:]
    monthday_list.append(monthday)
    
```

```python
# 연월/월일 컬럼 추가
raw['연월'] = yearmonth_list
raw['월일'] = monthday_list

raw.head()
```

```python
#월별 승객수 정리

# #1번 방식
# raw.pivot_table(index=, values, aggfunc=)

# # 2번 방식
# pd.raw.pivot_table(data= raw, index=, values, aggfunc=)

data_month = raw.pivot_table(index='연월', values='승차총승객수', aggfunc='sum').sort_values(by='승차총승객수', ascending=False)
data_month
```
