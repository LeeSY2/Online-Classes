```python
# 데이터 표 병합하기

# 첫번째 파일
fpath = './data/exam.xlsx'
A = pd.read_excel(fpath, index_col = '번호')
A
```

```python
# 두번쨰 파일
fpath = './data/exam_extra.xlsx'
B = pd.read_excel(fpath, index_col = '번호')
B
```

```python
# pd.merge(A,B, how ='left', left_on = 'A컬럼명' , left_index = True)
#                            right_on = 'B컬럼명' , right_index = True
```

```python
# 엑셀의 vloojup처럼 합치기
# A,B 테이블을 A 테이블에 있는 키 값을 기준으로 (how = 'left')
# A 테이블은 인덱스를 키값으로 , B 테이블은 인덱스를 키 값으로 병합

total = pd.merge(A,B,how='left', left_index = True, right_index = True)
total
```

```python
# A,B 테이블을 B 테이블에 있는 키 값을 기준으로 (how = 'right')
# A 테이블은 인덱스를 키값으로 , B 테이블은 인덱스를 키 값으로 병합

pd.merge(A,B,how='right', left_index = True, right_index = True)
```

```python
# A,B 테이블을 A,B 테이블 둘 다에 있는 키 값을 기준으로 (how = 'inner')
# A 테이블은 인덱스를 키값으로 , B 테이블은 인덱스를 키 값으로 병합

pd.merge(A,B,how='inner', left_index = True, right_index = True)
```

```python
# A,B 테이블을 A,B 테이블 둘 다에 있는 키 값을 기준으로 (how = 'outer')
# A 테이블은 인덱스를 키값으로 , B 테이블은 인덱스를 키 값으로 병합

pd.merge(A,B,how='outer', left_index = True, right_index = True)
```

```python
# 엑셀 파일로 저장
# df.excel('파일경로+파일명.xlsx', index=False)
```

```python
total = pd.merge(A,B,how='left', left_index = True, right_index = True)
total
```

```python
fpath = './data/exam_total.xlsx'
total.to_excel(fpath)
```

```python
# 판다스의 데이터프레임에서는 항상 인덱스 값을 가지낟
# 인덱스를 저장하고 싶지 않을 떄에늠 저장시에 index = False 옵션 쓰기

fpath = './data/exam_total_withoutindex.xlsx'
total.to_excel(fpath,index = False)
```

```python
import pandas as pd
fpath = './data/babyNamesUS.csv'
raw = pd.read_csv(fpath)
```

```python
raw.head()
```

```python
raw.info()
```

```python
# 집계하기
# df.pivot_table(index='컬럼명', columns='컬럼명',values='컬럼명',aggfunc='sum')
```

```python
# state, 성별, 출생연도에 상관없이 이름이 등록된 수를 합하여 정리
# 인덱스는 이름으로, 값은 등록된 수를 모두 더해서 피벗 테이블로 나타내기

raw.pivot_table(index='Name', values='Number', aggfunc='sum')
```

```python
# 앞서 생성된 데이터에서 성별 구분을 칼럼에 추가해서 피벗 테이블 만들기
# 인덱스는 이름이로, 값은 등록된 수의 합계, 컬럼은 성별로 구분하여 피벗테이블 만들기

name_df = raw.pivot_table(index='Name', values ='Number', aggfunc='sum', columns='Sex')
name_df
```

```python
name_df.info()
```

```python
# 비어있는 데이터 채워넣기

# 공통된 값 넣기(ex)0)
# 임의의 수 입력(평균 최소값 등)
# 비어있는 데이터 제외
```
