```python
import pandas as pd

file = './data/KBO_2019_player_gamestats.csv'
raw = pd.read_csv(file, encoding = 'cp949')

raw.head()
```

```python
# '상대' 컬럼에 어떤 값이 들어있는지 확인 : unique()


raw['상대'].unique()
```

```python
# @는 원정경기 , 없으면 홈그라운드
# '상대'컬럼에서 홈/원정 여부, 상대팀을 분리하여서 각각의 컬럼에 저장하기

opp_list = []
home_away_list = []

for opp in raw['상대'] : 
    
    if '@' in opp :
        home_away = '원정'
        opp = opp.replace('@', '')
        
    else:
        home_away ='홈'
        
    opp_list.append(opp)
    home_away_list.append(home_away)
        
raw['홈어웨이'] = home_away_list
raw['상대팀'] = opp_list

raw.head()
```

```python
home_away_list
```

```python
# 상대 팀별 실적 정리 위해 피벗 테이블 만들기
# ['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비']
# aggfunc 기본값은 mean 

data = raw.pivot_table( index = ['팀', '이름', '상대팀'],
               values = ['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
               aggfunc = 'sum')

data.head()
```

```python
# 상대팀별 타수가 0보다 큰 경우의 데이터만 선택

cond = data['타수'] > 0
data = data[cond]
```

```python
# resset_index()를 사용해 인덱스로 지정되어있는 팀/이름/생일을 컬럼으로 변경

data = data.reset_index()
```

```python
# 타자 주요 실적 계산 함수

def cal_hit(df):
    
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷']+df['사구'])/(df['타수'] + df['볼넷']+df['사구']+df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] / df['장타율']
    
    return df
```

```python
# 타자/상대 팀별 실적 계산 후 data에 추가하기

player_stats_opp= cal_hit(data)
```

```python
# 특정 팀을 상대로 강한 타자를 살피기
# 먼저 두산 팀을 상대로 타수가 10타수 보다 많은 선수중 출루율 상위 10명을 찾아보기


team = '두산'
cond = (player_stats_opp['상대팀'] == team) &  (player_stats_opp['타수'] > 10)

player_stats_opp[cond].sort_values(by = '출루율', ascending = False). head(10)
```
