```python
import pandas as pd

file='./data/babyNamesUS.csv'
raw=pd.read_csv(file)
```

```python
raw.head()
```

```python
raw.info()
```

```python
# 남여 상관없이 사용되는 공통된 이름은?

name_df = raw.pivot_table(index='Name', values = 'Number', aggfunc='sum', columns='Sex')
name_df

name_df= name_df.fillna(0)
name_df
```

```python
# 남여 비율 차이거 작을 수록 성별 구분이 없음
# 남여 이름 등록수 합계를 계산

name_df['Sum'] = name_df['M'] + name_df['F']
name_df
```

```python
# 남여 비율 계산


name_df['F_ratio'] = name_df['F']/name_df['Sum']
name_df['M_ratio'] = name_df['M']/name_df['Sum']
name_df.head()
```

```python
# 두비율의 차이 계산
name_df['F_ratio'] - name_df['M_ratio']
```

```python
# 절대값 전환
name_df['M_F_Gap'] = abs(name_df['F_ratio'] - name_df['M_ratio'])
name_df.head()
```

```python
# 이름이 가장 많이 사용한 순으로 내림차순 정렬

name_df = name_df.sort_values(by='Sum', ascending=False)
name_df.head()
```

```python
# 남여 비율의 차이가 0.1 보다 작은 경우 찾기

cond = name_df['M_F_Gap'] < 0.1

name_df[cond]
```

```python
# 남여 구분 없이 많이 쓰이는 이름 탑10

name_df[cond].head(10).index
```

```python
## 최근 트렌드에 따른 가장 대표적인 미국이름은?

raw['YearOfBirth'].unique()
```

```python
# 세대별로 그룹 만들기    
# 1930년대 이전, 1960년대 이전. 1990년대 이전, 2020년대 이전

year_class_list = []
for year in raw['YearOfBirth']:
    if year<= 1930:
        year_class = '1930년 이전'
    elif year<= 1960:
        year_class = '1960년 이전'
    elif year<= 1990:
        year_class = '1990년 이전'
    else :
        year_class = '2020년 이전'
    
    year_class_list.append(year_class)
```

```python
year_class_list
```

```python
# 세대 그룹명이 저낭된 리스트를 컬럼으로 추가

raw['year_class'] = year_class_list
raw.head()
```

```python
# pivot table 사용하여 세대별 이름 등록수 합계 표를 구하기

name_period = raw.pivot_table(index = ['Name','Sex'], values = 'Number', aggfunc = 'sum', columns = 'year_class')
name_period

name_period = name_period.fillna(0)    #결측치 채우기
name_period = name_period.astype(int)  #소수를 정수화
```

```python
# 전체 컬럼 합계 계산하기
#  1. 모든 컬럼을 하나씩 더하기
#  2. sum()이용하기 - 기본값이 axis=0으로 계산되고, row 합
```

```python
name_period.head()

name_period['1930년 이전'] + name_period['1960년 이전'] + name_period['1990년 이전'] + name_period['2020년 이전']
```

```python
# axis=1 이면 컬럼 합

name_period['sum'] = name_period.sum(axis=1)
name_period.head()
```

```python
# 모든 컬럼을 합계로 나누어 세대별 등록 비율을 계사하기
# 계산된 값은 기존칼럼 뒤애 '비율' 단어를 추가한 신규칼럼에 저장하기

for col in name_period.columns:
    
    col_new = col + '비율'
    name_period[col_new] =  name_period[col] / name_period['sum']
    
name_period.head()
```

```python
# 이름 사용 수 합계 2020년 이전 비율 , 1990년 이전 비율 기준으로  내림차순 으로 정렬

name_period = name_period.sort_values(by = ['sum', '2020년 이전비율', '1990년 이전비율'], ascending = False)
name_period
```

```python
# 인덱스가 여러 레벨로 되어있는 경우 인덱스를 활용해 컨트롤 하는 것은 복잡하기 때문에
# reset_index()를 이용해 인덱스로 설정된 이름과 성별을 컬럼으로 변경하기

name_period.index             # 이름과 성별로 되어있음
name_period = name_period.reset_index()     #기존 인덱스로 컬럼으로 변경
```

```python
# 남자이름만 선택해서 살피기

cond = name_period['Sex'] == 'M'
cond                                 # 시리즈 만들어짐

name_period[cond]                    # 조건에 맞는 행만 출력
```

```python
# 2020년 이전 비율이 30%보다 큰 경우의 이름은?
# 남자의 경우

cond_age = name_period['2020년 이전비율']>0.3
cond_sex = name_period['Sex'] =='M'
cond = cond_age & cond_sex

name_period[cond].head(5)
```

```python
# 여자인 경우

cond = name_period['Sex']=='F'
name_period[cond].head(5)
```

```python
# 2020년 이전 비율이  30%보다 큰 경우는?
# 여자인 경우

cond_age = name_period['2020년 이전비율'] > 0.3
cond_sex = name_period['Sex'] == 'F'
cond = cond_age & cond_sex

name_period[cond].head(5)
```
